#<h3> Task 2: Simulating BiasedCoin in Python</h3>

Flip the following biased coin 100 times. Calculate the total numbers of heads and tails, and then check the ratio of the number of heads and the number of tails.

$
BiasedCoin = \begin{array}{c|cc} \hookleftarrow & \mathbf{Head} & \mathbf{Tail} \\ \hline \mathbf{Head} & 0.6 & 0.6 \\  \mathbf{Tail} & 0.4 & 0.4  \end{array}
$


Do the same experiment 1000 times.

Do the same experiment 10,000 times.

Do the same experiment 100,000 times.

Do your results get close to the ideal case $ \mypar{ \dfrac{\mbox{\# of heads}}{\mbox{\# of tails}} = \dfrac{0.6}{0.4} = 1.50000000 } $?

```python
#
# you may use method 'randrange' for this task
# randrange(n) returns a value from {0,1,...,n-1} randomly
#

#import React, { useState, useEffect } from 'react';
import random
import matplotlib.pyplot as plt
import pandas as pd

def flip_biased_coin():
    """Flip a biased coin with P(Head) = 0.6, P(Tail) = 0.4"""
    return 'H' if random.random() < 0.6 else 'T'

def run_experiment(num_flips):
    """Run a coin flipping experiment with specified number of flips"""
    heads = 0
    tails = 0

    for _ in range(num_flips):
        if flip_biased_coin() == 'H':
            heads += 1
        else:
            tails += 1

    # Calculate ratio and statistics
    ratio = heads / tails if tails > 0 else heads
    expected_ratio = 0.6 / 0.4  # 1.5
    error = abs(ratio - expected_ratio)
    error_percentage = (error / expected_ratio) * 100

    return {
        'num_flips': num_flips,
        'heads': heads,
        'tails': tails,
        'ratio': ratio,
        'expected_ratio': expected_ratio,
        'error': error,
        'error_percentage': error_percentage
    }

def main():
    """Run all experiments and display results"""
    print("BIASED COIN SIMULATION")
    print("=" * 50)
    print("Coin Properties:")
    print("P(Head) = 0.6")
    print("P(Tail) = 0.4")
    print("Expected Ratio (Heads/Tails) = 0.6/0.4 = 1.5")
    print("=" * 50)

    # Run experiments with different numbers of flips
    experiments = [100, 1000, 10000, 100000]
    results = []

    print("\nRUNNING EXPERIMENTS...")
    print("-" * 50)

    for num_flips in experiments:
        print(f"Running experiment with {num_flips:,} flips...")
        result = run_experiment(num_flips)
        results.append(result)

        print(f"  Heads: {result['heads']:,}")
        print(f"  Tails: {result['tails']:,}")
        print(f"  Actual Ratio: {result['ratio']:.6f}")
        print(f"  Expected Ratio: {result['expected_ratio']:.6f}")
        print(f"  Error: {result['error']:.6f}")
        print(f"  Error Percentage: {result['error_percentage']:.2f}%")
        print()

    # Create summary table
    print("\nSUMMARY TABLE")
    print("-" * 80)
    print(f"{'Flips':>10} {'Heads':>8} {'Tails':>8} {'Actual Ratio':>12} {'Expected':>10} {'Error':>8} {'Error %':>8}")
    print("-" * 80)

    for result in results:
        print(f"{result['num_flips']:>10,} {result['heads']:>8,} {result['tails']:>8,} "
              f"{result['ratio']:>12.6f} {result['expected_ratio']:>10.6f} "
              f"{result['error']:>8.6f} {result['error_percentage']:>7.2f}%")

    # Create visualizations
    create_plots(results)

    # Analysis
    print("\n" + "=" * 50)
    print("ANALYSIS & CONCLUSIONS")
    print("=" * 50)
    print("1. LAW OF LARGE NUMBERS:")
    print("   As the number of flips increases, the actual ratio")
    print("   converges toward the expected value of 1.5")
    print()
    print("2. ERROR REDUCTION:")
    for result in results:
        print(f"   {result['num_flips']:,} flips: {result['error_percentage']:.2f}% error")
    print()
    print("3. STATISTICAL SIGNIFICANCE:")
    print("   With 100,000 flips, we get very close to the theoretical")
    print("   ratio, confirming our biased coin works as expected.")
    print()
    print("4. The results demonstrate that probability theory works")
    print("   beautifully in practice!")

def create_plots(results):
    """Create visualizations of the results"""
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

    # Plot 1: Heads vs Tails bar chart
    trials = [f"{r['num_flips']:,}" for r in results]
    heads = [r['heads'] for r in results]
    tails = [r['tails'] for r in results]

    x = range(len(trials))
    width = 0.35

    ax1.bar([i - width/2 for i in x], heads, width, label='Heads', color='skyblue')
    ax1.bar([i + width/2 for i in x], tails, width, label='Tails', color='lightcoral')
    ax1.set_xlabel('Number of Trials')
    ax1.set_ylabel('Count')
    ax1.set_title('Heads vs Tails Distribution')
    ax1.set_xticks(x)
    ax1.set_xticklabels(trials, rotation=45)
    ax1.legend()
    ax1.grid(axis='y', alpha=0.3)

    # Plot 2: Ratio convergence
    num_flips = [r['num_flips'] for r in results]
    actual_ratios = [r['ratio'] for r in results]
    expected_line = [1.5] * len(results)

    ax2.semilogx(num_flips, actual_ratios, 'bo-', linewidth=2, markersize=8, label='Actual Ratio')
    ax2.semilogx(num_flips, expected_line, 'r--', linewidth=2, label='Expected Ratio (1.5)')
    ax2.set_xlabel('Number of Flips (log scale)')
    ax2.set_ylabel('Heads/Tails Ratio')
    ax2.set_title('Convergence to Expected Ratio')
    ax2.legend()
    ax2.grid(alpha=0.3)
    ax2.set_ylim(1.0, 2.0)

    # Plot 3: Error percentage
    error_percentages = [r['error_percentage'] for r in results]

    ax3.semilogx(num_flips, error_percentages, 'go-', linewidth=2, markersize=8)
    ax3.set_xlabel('Number of Flips (log scale)')
    ax3.set_ylabel('Error Percentage (%)')
    ax3.set_title('Error Reduction with Sample Size')
    ax3.grid(alpha=0.3)

    # Plot 4: Probability comparison
    theoretical_heads = [n * 0.6 for n in num_flips]
    theoretical_tails = [n * 0.4 for n in num_flips]
    actual_heads = heads
    actual_tails = tails

    ax4.scatter(theoretical_heads, actual_heads, color='blue', alpha=0.7, s=100, label='Heads')
    ax4.scatter(theoretical_tails, actual_tails, color='red', alpha=0.7, s=100, label='Tails')

    # Perfect correlation line
    max_val = max(max(theoretical_heads), max(actual_heads))
    ax4.plot([0, max_val], [0, max_val], 'k--', alpha=0.5, label='Perfect Correlation')

    ax4.set_xlabel('Theoretical Count')
    ax4.set_ylabel('Actual Count')
    ax4.set_title('Theoretical vs Actual Results')
    ax4.legend()
    ax4.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Set random seed for reproducibility (optional)
    # random.seed(42)  # Uncomment to get same results each time

    main()
#
```

<h3> Extra: Programming a biased coin </h3>

We use a simple method to create a biased coin.

First, we pick a range for the precision of probabilities, say $ N $, as $ N = 11, 101, 1001, \mbox{ or }, 10^k+1 $ for some $ k > 3 $.

Second, we pick the bias, say $ B $, as an integer in $ \{0,\ldots,N\} $.

We fix $ N $ and $ B $.

Third, we pick a random integer in $ \{0,1,\ldots,N-1\} $:

- if it is less than $ B $, we output "Heads" and
- if it is equal to or greater than $ B $, we output "Tails"

    
In this way, we have a biased coin "landing on" heads with probability $ \frac{B}{N} $ including 0 and 1.

Remark that we pick $ N = 10^k+1 $ as an odd number. In this way, the coin cannot be fair as long as $ B $ is an integer. Because, the half of an odd integer is not an integer.
